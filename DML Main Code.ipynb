```python
# Loading necessary libraries
import pandas as pd              #Data manipulation and analysis
import numpy as np              #Reading numerical data
from sklearn.preprocessing import StandardScaler  # For scaling the features 
from sklearn.model_selection import (
    train_test_split,          # Splitting the dataset for testing and training 
    GridSearchCV,              # GridSearch for Hyperparameter tuning
    learning_curve            # Analysing model learning behavior 
)
from sklearn.cluster import KMeans  # For clustering analysis
from sklearn.tree import DecisionTreeClassifier, plot_tree  # For decision tree model 
from sklearn.linear_model import LinearRegression  # For linear regression  
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors  # For KNN model
from sklearn.metrics import (
    confusion_matrix,          # For evaluating metrics
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,                # For ROC curve analysis
    auc,
    classification_report,
    mean_squared_error,
    r2_score,
    silhouette_score  # Added for cluster evaluation
)

# K-Means- SANDEEP
# Linear Regression- REET SHARMA
# DBSCAN- ABHISHEK BASAVARAJ
# Decision Tree- All team Members
# KNN- OMANA
# MLP- ABDULLAHI
# CNN- ABDULLAHI & SANDEEP
# Every Team member contributed in all the Algorithm


from scipy import stats        # For Statistical Analysis
import matplotlib.pyplot as plt  # For plotting the graphs for analysis
import seaborn as sns          # For enhanced visualizations
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from sklearn.neural_network import MLPClassifier
from sklearn.cluster import DBSCAN

class SpotifyAnalyzer:
    def __init__(self):
        """Initialize the SpotifyAnalyzer with feature definitions and scaler"""
        self.features = [
            'danceability', 'energy', 'key', 'loudness', 'mode',
            'speechiness', 'acousticness', 'instrumentalness',
            'liveness', 'valence', 'tempo', 'duration_ms'
        ]
        self.scaler = StandardScaler()

    def prepare_data(self, data):
        """Prepare the dataset for analysis"""
        X = data[self.features]
        y = data['track_popularity']
        X_scaled = self.scaler.fit_transform(X)
        return X_scaled, y

    def optimize_dbscan_params(self, X):
        """
        Optimize DBSCAN parameters using nearest neighbors analysis

        Parameters:
        -----------
        X : array-like
            Standardized feature matrix

        Returns:
        --------
        tuple : (eps, min_samples)
            Optimized DBSCAN parameters
        """
        neigh = NearestNeighbors(n_neighbors=2)
        nbrs = neigh.fit(X)
        distances, _ = nbrs.kneighbors(X)
        distances = np.sort(distances[:, 1])

        acceleration = np.diff(distances, 2)
        elbow_index = np.argmax(acceleration) + 1
        eps = distances[elbow_index]

        min_samples = X.shape[1] * 2

        return eps, min_samples

    def perform_clustering(self, X, k=5):
        """
        Perform both K-means and DBSCAN clustering

        Parameters:
        -----------
        X : array-like
            Standardized feature matrix
        k : int, default=5
            Number of clusters for K-means

        Returns:
        --------
        tuple : (kmeans_labels, clustering_results)
            - kmeans_labels: Labels for model training
            - clustering_results: Dictionary containing both algorithms' outputs
        """
        clustering_results = {}
        
        # THIS WAS PERFORMED BY SANDEEP [K-means]
        
        # 1. K-means clustering
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans_labels = kmeans.fit_predict(X)

        # Map K-means clusters to popularity levels
        kmeans_popularity = {}
        for cluster in range(k):
            mask = kmeans_labels == cluster
            kmeans_popularity[cluster] = np.mean(self.original_y[mask])

        # Create ordered K-means labels
        kmeans_new_labels = np.zeros_like(kmeans_labels)
        for new_label, (cluster, _) in enumerate(
            sorted(kmeans_popularity.items(), key=lambda x: x[1])):
            kmeans_new_labels[kmeans_labels == cluster] = new_label
            
         #THIS WAS PERFORMED BY ABHISHEK
        
        # 2. DBSCAN clustering
        eps, min_samples = self.optimize_dbscan_params(X)
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan_labels = dbscan.fit_predict(X)

        # Map DBSCAN clusters to popularity levels
        unique_clusters = np.unique(dbscan_labels)
        dbscan_popularity = {}
        for cluster in unique_clusters:
            mask = dbscan_labels == cluster
            dbscan_popularity[cluster] = np.mean(self.original_y[mask])

        # Create ordered DBSCAN labels
        dbscan_new_labels = np.zeros_like(dbscan_labels)
        for new_label, (cluster, _) in enumerate(
            sorted(dbscan_popularity.items(), key=lambda x: x[1])):
            dbscan_new_labels[dbscan_labels == cluster] = new_label

        # Store results
        clustering_results['kmeans'] = {
            'labels': kmeans_new_labels,
            'centroids': kmeans.cluster_centers_,
            'silhouette': silhouette_score(X, kmeans_labels),
            'n_clusters': k
        }

        clustering_results['dbscan'] = {
            'labels': dbscan_new_labels,
            'eps': eps,
            'min_samples': min_samples,
            'silhouette': silhouette_score(X, dbscan_labels) if len(unique_clusters) > 1 else 0,
            'n_clusters': len(unique_clusters[unique_clusters != -1]),
            'noise_points': np.sum(dbscan_labels == -1)
        }

        return kmeans_new_labels, clustering_results

    def build_mlp_classifier(self):
        """Build and configure the MLP classifier"""
        return MLPClassifier(
            hidden_layer_sizes=(100, 50),
            activation='relu',
            solver='adam',
            alpha=0.0001,
            batch_size='auto',
            learning_rate='adaptive',
            max_iter=1000,
            early_stopping=True,
            random_state=42
        )

    def build_cnn_model(self, input_shape, num_classes):
        """
        Build and compile the CNN model

        Parameters:
        -----------
        input_shape : tuple
            Shape of input data (features, 1)
        num_classes : int
            Number of output classes

        Returns:
        --------
        model : keras.Model
            Compiled CNN model
        """
        model = Sequential([
            # Reshape input for CNN
            Conv1D(filters=32, kernel_size=3, activation='relu',
                  input_shape=input_shape, padding='same'),
            MaxPooling1D(pool_size=2),

            Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
            MaxPooling1D(pool_size=2),

            Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),

            Flatten(),
            Dense(128, activation='relu'),
            Dropout(0.5),
            Dense(64, activation='relu'),
            Dense(num_classes, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def prepare_cnn_data(self, X):
        """Reshape data for CNN input"""
        return X.reshape(X.shape[0], X.shape[1], 1)

    def tune_decision_tree(self, X, y):
        """
        Tune Decision Tree hyperparameters using GridSearchCV

        Parameters:
        -----------
        X : array-like
            Training features
        y : array-like
            Target labels

        Returns:
        --------
        tuple : (best_model, best_params)
            - best_model: Optimized Decision Tree model
            - best_params: Best hyperparameters found
        """
        # Define hyperparameter search space
        param_grid = {
            'criterion': ['gini', 'entropy'],     # Splitting criteria
            'splitter': ['best', 'random'],       # Split selection strategy
            'max_depth': [3, 5, 7, 10, None],     # Maximum tree depth
            'min_samples_split': [2, 5, 10],      # Min samples for splitting
            'min_samples_leaf': [1, 2, 4],        # Min samples at leaf
            'max_features': ['sqrt', 'log2', None],# Feature selection for splits
            'ccp_alpha': [0.0, 0.01, 0.02, 0.03]  # Pruning parameter
        }

        # Initialize model and perform grid search
        dt = DecisionTreeClassifier(random_state=42)
        grid_search = GridSearchCV(dt, param_grid, cv=5,
                                 scoring='accuracy', n_jobs=-1)
        grid_search.fit(X, y)

        return grid_search.best_estimator_, grid_search.best_params_
    def tune_knn(self, X, y):
        """
        Tune K-Nearest Neighbors hyperparameters using GridSearchCV

        Parameters:
        -----------
        X : array-like
            Training features
        y : array-like
            Target labels

        Returns:
        --------
        tuple : (best_model, best_params)
            - best_model: Optimized KNN model
            - best_params: Best hyperparameters found
        """
        # Defining parameter grid for KNN optimization
        param_grid = {
            'n_neighbors': [3, 5, 7, 9, 11],      # Number of neighbors to consider
            'weights': ['uniform', 'distance'],    # Weight function for prediction
            'metric': ['euclidean', 'manhattan', 'minkowski'],  # Distance metric
            'p': [1, 2]  # Power parameter for Minkowski metric (1=manhattan, 2=euclidean)
        }

        # Initializing and performing grid search
        knn = KNeighborsClassifier()
        grid_search = GridSearchCV(knn, param_grid, cv=5,
                                 scoring='accuracy', n_jobs=-1)
        grid_search.fit(X, y)

        return grid_search.best_estimator_, grid_search.best_params_

    def check_model_learning(self, model, X, y, model_name):
        """
        Generate learning curves to check for overfitting/underfitting

        Parameters:
        -----------
        model : estimator object
            The machine learning model
        X : array-like
            Training features
        y : array-like
            Target labels
        model_name : str
            Name of the model for plotting
        """
        # Generating Learning curves with cross-validation
        train_sizes, train_scores, val_scores = learning_curve(
            model, X, y, cv=5, n_jobs=-1,
            train_sizes=np.linspace(0.1, 1.0, 10)  # Training size points from 10% to 100%
        )

        # Creating visualization
        plt.figure(figsize=(10, 6))
        # Plotting training scores
        plt.plot(train_sizes, np.mean(train_scores, axis=1), 'r-',
                label='Training score')
        # Plotting validation scores
        plt.plot(train_sizes, np.mean(val_scores, axis=1), 'b-',
                label='Cross-validation score')

        # Adding confidence intervals
        plt.fill_between(train_sizes,
                        np.mean(train_scores, axis=1) - np.std(train_scores, axis=1),
                        np.mean(train_scores, axis=1) + np.std(train_scores, axis=1),
                        alpha=0.1, color='r')
        plt.fill_between(train_sizes,
                        np.mean(val_scores, axis=1) - np.std(val_scores, axis=1),
                        np.mean(val_scores, axis=1) + np.std(val_scores, axis=1),
                        alpha=0.1, color='b')

        # Customizing plot
        plt.xlabel('Training Examples')
        plt.ylabel('Score')
        plt.title(f'Learning Curves ({model_name})')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

    def analyze_correlations(self, data):
        """
        Analyze and visualize correlations between features and target

        Parameters:
        -----------
        data : pandas.DataFrame
            Input dataset containing features and target variable

        Returns:
        --------
        dict : Dictionary containing correlation analysis results
        """
        # Creating correlation matrix including target variable
        features_with_target = self.features + ['track_popularity']
        correlation_matrix = data[features_with_target].corr()

        # Calculating feature correlations with target
        target_correlations = correlation_matrix['track_popularity'].sort_values(ascending=False)

        # Identifying strong correlations between features
        feature_correlations = []
        for i in range(len(self.features)):
            for j in range(i + 1, len(self.features)):
                corr = correlation_matrix.iloc[i, j]
                if abs(corr) > 0.5:  # Threshold for strong correlation
                    feature_correlations.append({
                        'feature1': self.features[i],
                        'feature2': self.features[j],
                        'correlation': corr
                    })

        # Visualizing correlation matrix
        plt.figure(figsize=(12, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
        plt.title('Feature Correlation Matrix')
        plt.tight_layout()
        plt.show()

        # Visualizing feature correlations with target
        plt.figure(figsize=(10, 6))
        target_correlations[:-1].plot(kind='bar')  # Exclude self-correlation
        plt.title('Feature Correlations with Track Popularity')
        plt.xlabel('Features')
        plt.ylabel('Correlation Coefficient')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Analyzing feature interactions
        significant_pairs = []
        for pair in feature_correlations:
            if abs(pair['correlation']) > 0.7:  # Threshold for significant correlation
                significant_pairs.append(pair)

                # Visualizing relationship for highly correlated features
                plt.figure(figsize=(8, 6))
                sns.scatterplot(data=data, x=pair['feature1'], y=pair['feature2'],
                              hue='track_popularity', palette='viridis')
                plt.title(f"Relationship: {pair['feature1']} vs {pair['feature2']}")
                plt.xlabel(pair['feature1'])
                plt.ylabel(pair['feature2'])
                plt.tight_layout()
                plt.show()

        return {
            'correlation_matrix': correlation_matrix,
            'target_correlations': target_correlations,
            'feature_correlations': feature_correlations,
            'significant_pairs': significant_pairs
        }

    def analyze_and_predict(self, data):
        """Complete analysis pipeline: preprocessing, model training, and evaluation"""
        # Performing correlation analysis first
        print("Performing correlation analysis...")
        correlation_results = self.analyze_correlations(data)

        # Printing correlation insights
        print("\nCorrelation Analysis Insights:")
        print("Top 3 features positively correlated with popularity:")
        top_positive = correlation_results['target_correlations'][:-1].nlargest(3)
        for feature, corr in top_positive.items():
            print(f"{feature}: {corr:.3f}")

        print("\nTop 3 features negatively correlated with popularity:")
        top_negative = correlation_results['target_correlations'][:-1].nsmallest(3)
        for feature, corr in top_negative.items():
            print(f"{feature}: {corr:.3f}")

        if correlation_results['significant_pairs']:
            print("\nHighly correlated feature pairs:")
            for pair in correlation_results['significant_pairs']:
                print(f"{pair['feature1']} & {pair['feature2']}: {pair['correlation']:.3f}")

        # Continuing with existing pipeline
        X_scaled, y = self.prepare_data(data)
        self.original_y = y

        print("\nPerforming clustering...")
        cluster_labels, clustering_results = self.perform_clustering(X_scaled)

        # Splitting data
        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, cluster_labels, test_size=0.2, random_state=42
        )

        # Preparing data for CNN
        X_train_cnn = self.prepare_cnn_data(X_train)
        X_test_cnn = self.prepare_cnn_data(X_test)
        y_train_cat = to_categorical(y_train)
        y_test_cat = to_categorical(y_test)

        # THIS ALGORITHM WAS DONE BY REET
        
        # Training existing models
        print("\nTraining Linear Regression...")
        lr_model = LinearRegression()
        lr_model.fit(X_train, y_train)
        lr_pred = np.round(lr_model.predict(X_test))
        
        #EVERYONE TOGETHER AS A GROUP WORKED ON THIS SECTION
        
        print("\nTraining Decision Tree...")
        dt_model, dt_params = self.tune_decision_tree(X_train, y_train)
        dt_pred = dt_model.predict(X_test)
        
        #THIS WAS PERFORMED BY OMANA
        
        print("\nTraining KNN...")
        knn_model, knn_params = self.tune_knn(X_train, y_train)
        knn_pred = knn_model.predict(X_test)

        #THIS WAS PERFORMED BY ABDULLAHI
        
        # Training MLP
        print("\nTraining MLP...")
        mlp_model = self.build_mlp_classifier()
        mlp_model.fit(X_train, y_train)
        mlp_pred = mlp_model.predict(X_test)

        #THIS WAS PERFORMED BY SANDEEP & ABDULLAHI
        
        # Training CNN
        print("\nTraining CNN...")
        num_classes = len(np.unique(y_train))
        cnn_model = self.build_cnn_model((X_train.shape[1], 1), num_classes)

        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )

        cnn_history = cnn_model.fit(
            X_train_cnn, y_train_cat,
            epochs=50,
            batch_size=32,
            validation_split=0.2,
            callbacks=[early_stopping],
            verbose=1
        )

        # Storing history in the model for later access
        cnn_model.history = cnn_history
        cnn_pred = np.argmax(cnn_model.predict(X_test_cnn), axis=1)

        # Storing results for all models
        results = {
            'Decision Tree': {
                'accuracy': accuracy_score(y_test, dt_pred),
                'confusion_matrix': confusion_matrix(y_test, dt_pred),
                'params': dt_params,
                'predictions': dt_pred
            },
            'KNN': {
                'accuracy': accuracy_score(y_test, knn_pred),
                'confusion_matrix': confusion_matrix(y_test, knn_pred),
                'params': knn_params,
                'predictions': knn_pred
            },
            'Linear Regression': {
                'accuracy': accuracy_score(y_test, lr_pred),
                'confusion_matrix': confusion_matrix(y_test, lr_pred),
                'mse': mean_squared_error(y_test, lr_pred),
                'predictions': lr_pred,
                'params': 'N/A'
            },
            'MLP': {
                'accuracy': accuracy_score(y_test, mlp_pred),
                'confusion_matrix': confusion_matrix(y_test, mlp_pred),
                'predictions': mlp_pred,
                'params': mlp_model.get_params()
            },
            'CNN': {
                'accuracy': accuracy_score(y_test, cnn_pred),
                'confusion_matrix': confusion_matrix(y_test, cnn_pred),
                'predictions': cnn_pred,
                'history': cnn_history.history  # Store history directly in results
            }
        }

        # Storing models
        models = {
            'Decision Tree': dt_model,
            'KNN': knn_model,
            'Linear Regression': lr_model,
            'MLP': mlp_model,
            'CNN': cnn_model
        }

        # Storing parameters
        params = {
            'X_train': X_train,
            'X_test': X_test,
            'X_train_cnn': X_train_cnn,
            'X_test_cnn': X_test_cnn,
            'y_train': y_train,
            'y_test': y_test,
            'y_train_cat': y_train_cat,
            'y_test_cat': y_test_cat
        }

        self.plot_results(results, X_scaled, clustering_results,
                         clustering_results['kmeans']['centroids'])

        return results, models, params

    def plot_neural_network_history(self, history):
        """
        Plot training history for neural networks

        Parameters:
        -----------
        history : keras.callbacks.History
            Training history from CNN model
        """
        plt.figure(figsize=(12, 4))

        # Plotting accuracy
        plt.subplot(121)
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()

        # Plotting loss
        plt.subplot(122)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.tight_layout()
        plt.show()

    def analyze_neural_networks(self, model, X_test, y_test, results):
        """
        Analyze neural network model performance

        Parameters:
        -----------
        model : Union[MLPClassifier, keras.Model]
            Trained neural network model
        X_test : array-like
            Test features
        y_test : array-like
            Test labels
        results : dict
            Dictionary to store analysis results
        """
        if isinstance(model, MLPClassifier):
            # MLP-specific analysis
            results['loss_curve'] = model.loss_curve_

            plt.figure(figsize=(10, 5))
            plt.plot(model.loss_curve_)
            plt.title('MLP Loss Curve')
            plt.xlabel('Iteration')
            plt.ylabel('Loss')
            plt.show()
        else:
            # CNN-specific analysis - access history directly from the CNN model
            try:
                history = model.history.history
                plt.figure(figsize=(12, 4))

                # Plotting accuracy
                plt.subplot(121)
                plt.plot(history['accuracy'], label='Training Accuracy')
                plt.plot(history['val_accuracy'], label='Validation Accuracy')
                plt.title('CNN Accuracy')
                plt.xlabel('Epoch')
                plt.ylabel('Accuracy')
                plt.legend()

                # Plotting loss
                plt.subplot(122)
                plt.plot(history['loss'], label='Training Loss')
                plt.plot(history['val_loss'], label='Validation Loss')
                plt.title('CNN Loss')
                plt.xlabel('Epoch')
                plt.ylabel('Loss')
                plt.legend()

                plt.tight_layout()
                plt.show()
            except (AttributeError, KeyError):
                print("Training history not available for CNN model")

    def plot_results(self, results, X, cluster_results, centroids):
        """Visualize clustering results and model performance"""
        plt.figure(figsize=(20, 15))

        # Plotting K-means clustering results
        plt.subplot(331)
        plt.scatter(X[:, 0], X[:, 1], c=cluster_results['kmeans']['labels'], cmap='viridis')
        plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='r')
        plt.title(f'K-means Clustering\nSilhouette Score: {cluster_results["kmeans"]["silhouette"]:.3f}')

        # Plotting DBSCAN clustering results
        plt.subplot(332)
        plt.scatter(X[:, 0], X[:, 1], c=cluster_results['dbscan']['labels'], cmap='viridis')
        plt.title(f'DBSCAN Clustering\nSilhouette Score: {cluster_results["dbscan"]["silhouette"]:.3f}\n'
                 f'Clusters: {cluster_results["dbscan"]["n_clusters"]}, '
                 f'Noise Points: {cluster_results["dbscan"]["noise_points"]}')

        # Plotting confusion matrices for each model
        for i, (name, result) in enumerate(results.items(), 1):
            plt.subplot(334 + i)
            sns.heatmap(result['confusion_matrix'], annot=True, fmt='d', cmap='Blues')
            plt.title(f'{name}\nAccuracy: {result["accuracy"]:.3f}')

        plt.tight_layout()
        plt.show()


    def analyze_results_in_detail(self, results, models, X_train, X_test, y_train, y_test):
        """
        Perform detailed analysis of model performance

        Parameters:
        -----------
        results : dict
            Dictionary containing model results
        models : dict
            Dictionary of trained models
        X_train, X_test, y_train, y_test : array-like
            Training and test data splits

        Returns:
        --------
        dict : Detailed results for each model
        """
        detailed_results = {}

        for name, model in models.items():
            print(f"\nDetailed Analysis for {name}:")

            y_pred = results[name]['predictions']
            detailed_results[name] = {
                'accuracy': accuracy_score(y_test, y_pred),
                'precision': precision_score(y_test, y_pred, average='weighted'),
                'recall': recall_score(y_test, y_pred, average='weighted'),
                'f1': f1_score(y_test, y_pred, average='weighted'),
                'confusion_matrix': confusion_matrix(y_test, y_pred)
            }

            # Adding neural network analysis
            if name in ['MLP', 'CNN']:
                self.analyze_neural_networks(model, X_test, y_test, detailed_results[name])

            # Performing model-specific analysis
            if name == 'Decision Tree':
                self.analyze_decision_tree(model, X_train, y_train, detailed_results[name])
            elif name == 'KNN':
                self.analyze_knn(model, X_test, y_test, detailed_results[name])
            elif name == 'Linear Regression':
                self.analyze_linear_regression(model, X_test, y_test, detailed_results[name])

            # Printing classification report
            print("\nClassification Report:")
            print(classification_report(y_test, y_pred))

            # Generating visualizations
            self.plot_model_analysis(name, model, X_test, y_test, y_pred, detailed_results[name])

        # Comparing models
        self.plot_model_comparison(detailed_results)

        return detailed_results

    def analyze_decision_tree(self, model, X_train, y_train, results):
        """
        Analyze Decision Tree model characteristics

        Parameters:
        -----------
        model : DecisionTreeClassifier
            Trained decision tree model
        X_train : array-like
            Training features
        y_train : array-like
            Training labels
        results : dict
            Dictionary to store analysis results
        """
        # Calculating feature importance
        results['feature_importance'] = dict(zip(self.features,
                                               model.feature_importances_))

        # Analyzing tree complexity
        results['tree_depth'] = model.get_depth()
        results['n_leaves'] = model.get_n_leaves()

        # Calculating training score
        results['train_score'] = model.score(X_train, y_train)

        # Visualizing tree structure
        plt.figure(figsize=(20,10))
        plot_tree(model, feature_names=self.features,
                 class_names=[str(i) for i in range(5)],
                 filled=True, rounded=True)
        plt.title("Decision Tree Structure")
        plt.show()

    def analyze_knn(self, model, X_test, y_test, results):
        """
        Analyze K-Nearest Neighbors model characteristics

        Parameters:
        -----------
        model : KNeighborsClassifier
            Trained KNN model
        X_test : array-like
            Test features
        y_test : array-like
            Test labels
        results : dict
            Dictionary to store analysis results
        """
        # Calculating neighbor distances
        distances, indices = model.kneighbors(X_test)
        results['avg_neighbor_distance'] = np.mean(distances, axis=1)
        results['std_neighbor_distance'] = np.std(distances, axis=1)

        # Plotting neighbor distance distribution
        plt.figure(figsize=(10,5))
        plt.hist(results['avg_neighbor_distance'], bins=50)
        plt.title("Distribution of Average Neighbor Distances")
        plt.xlabel("Average Distance")
        plt.ylabel("Frequency")
        plt.show()

    def analyze_linear_regression(self, model, X_test, y_test, results):
        """
        Analyze Linear Regression model performance

        Parameters:
        -----------
        model : LinearRegression
            Trained linear regression model
        X_test : array-like
            Test features
        y_test : array-like
            Test labels
        results : dict
            Dictionary to store analysis results
        """
        # Making predictions and calculate metrics
        y_pred = model.predict(X_test)
        results['mse'] = mean_squared_error(y_test, y_pred)
        results['rmse'] = np.sqrt(results['mse'])
        results['r2'] = r2_score(y_test, y_pred)

        # Performing residual analysis
        residuals = y_test - y_pred
        results['residuals_mean'] = np.mean(residuals)
        results['residuals_std'] = np.std(residuals)

        # Testing residuals for normality
        _, p_value = stats.normaltest(residuals)
        results['residuals_normality_p_value'] = p_value
    def plot_model_analysis(self, name, model, X_test, y_test, y_pred, results):
        """
        Generate comprehensive visualization of model performance

        Parameters:
        -----------
        name : str
            Model name
        model : estimator object
            Trained model
        X_test : array-like
            Test features
        y_test : array-like
            True labels
        y_pred : array-like
            Predicted labels
        results : dict
            Model results dictionary
        """
        plt.figure(figsize=(20, 10))

        # 1. Confusion Matrix Visualization
        plt.subplot(231)
        sns.heatmap(results['confusion_matrix'], annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusion Matrix - {name}')

        # 2. ROC Curve (for models with probability predictions)
        if hasattr(model, "predict_proba"):
            plt.subplot(232)
            y_prob = model.predict_proba(X_test)
            # Plot ROC curve for each class
            for i in range(5):
                fpr, tpr, _ = roc_curve(y_test == i, y_prob[:, i])
                plt.plot(fpr, tpr, label=f'Class {i} (AUC = {auc(fpr, tpr):.2f})')
            plt.plot([0, 1], [0, 1], 'k--')  # Reference line
            plt.xlabel('False Positive Rate')
            plt.ylabel('True Positive Rate')
            plt.title('ROC Curves')
            plt.legend()

        # 3. Error Distribution Analysis
        plt.subplot(233)
        errors = y_test - y_pred
        plt.hist(errors, bins=20)
        plt.title('Prediction Error Distribution')
        plt.xlabel('Error')
        plt.ylabel('Frequency')

        # 4. Feature Importance (if available)
        if 'feature_importance' in results:
            plt.subplot(234)
            importance_df = pd.DataFrame(list(results['feature_importance'].items()),
                                       columns=['Feature', 'Importance'])
            importance_df = importance_df.sort_values('Importance', ascending=False)
            sns.barplot(x='Importance', y='Feature', data=importance_df)
            plt.title('Feature Importance')

        # 5. Model-specific plots
        if name == 'KNN':
            # KNN-specific: Neighbor distance analysis
            plt.subplot(235)
            plt.scatter(range(len(results['avg_neighbor_distance'])),
                       results['avg_neighbor_distance'])
            plt.title('Average Neighbor Distance per Sample')
            plt.xlabel('Sample Index')
            plt.ylabel('Average Distance')
        elif name == 'Linear Regression':
            # Linear Regression-specific: Predicted vs Actual
            plt.subplot(235)
            plt.scatter(y_pred, y_test)
            plt.plot([y_test.min(), y_test.max()],
                    [y_test.min(), y_test.max()], 'r--')
            plt.title('Predicted vs Actual Values')
            plt.xlabel('Predicted')
            plt.ylabel('Actual')

        plt.tight_layout()
        plt.show()


    def plot_model_comparison(self, detailed_results):
        """
        Compare performance metrics across all models

        Parameters:
        -----------
        detailed_results : dict
            Dictionary containing detailed results for each model
        """
        plt.figure(figsize=(15, 10))

        # Plotting comparison of different metrics
        metrics = ['accuracy', 'precision', 'recall', 'f1']
        for i, metric in enumerate(metrics, 1):
            plt.subplot(2, 2, i)
            values = [results[metric] for results in detailed_results.values()]
            plt.bar(detailed_results.keys(), values)
            plt.title(f'{metric.capitalize()} Comparison')
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()

    # Every Team member contributed in all the Algorithm

def main():
    """
    Main execution function for the Spotify song analysis
    """
    # Loading and prepare data
    data = pd.read_csv('/Users/sandeep/Downloads/spotify_songs.csv')
    analyzer = SpotifyAnalyzer()

    # Performing initial analysis and model training
    print("Starting analysis pipeline...")
    results, models, data_split = analyzer.analyze_and_predict(data)

    # Performing detailed analysis of results
    print("\nPerforming detailed analysis...")
    detailed_results = analyzer.analyze_results_in_detail(
        results, models, data_split['X_train'],
        data_split['X_test'],
        data_split['y_train'],
        data_split['y_test'])

    # Printing comprehensive summary
    print("\nComprehensive Analysis Summary:")
    for model_name, results in detailed_results.items():
        # Printing basic metrics for all models
        print(f"\n{model_name} Results:")
        print("=====================================")
        print(f"Accuracy: {results['accuracy']:.3f}")
        print(f"Precision: {results['precision']:.3f}")
        print(f"Recall: {results['recall']:.3f}")
        print(f"F1 Score: {results['f1']:.3f}")

        # Printing model-specific metrics
        if model_name == 'Decision Tree':
            print(f"\nTree Complexity:")
            print(f"Depth: {results['tree_depth']}")
            print(f"Number of Leaves: {results['n_leaves']}")
            print("\nTop 5 Important Features:")
            sorted_features = sorted(results['feature_importance'].items(),
                                   key=lambda x: x[1], reverse=True)[:5]
            for feature, importance in sorted_features:
                print(f"{feature}: {importance:.3f}")

        elif model_name == 'KNN':
            print(f"\nNeighbor Distance Statistics:")
            print(f"Average Distance: {np.mean(results['avg_neighbor_distance']):.3f}")
            print(f"Std Dev Distance: {np.mean(results['std_neighbor_distance']):.3f}")

        elif model_name == 'Linear Regression':
            print(f"\nRegression Metrics:")
            print(f"RMSE: {results['rmse']:.3f}")
            print(f"R Score: {results['r2']:.3f}")
            print(f"Residuals Normality p-value: {results['residuals_normality_p_value']:.3f}")

    # Identifying and report best model
    best_model_name = max(detailed_results.items(),
                         key=lambda x: x[1]['accuracy'])[0]
    best_model = models[best_model_name]

    print(f"\nBest performing model: {best_model_name}")
    print(f"Accuracy: {detailed_results[best_model_name]['accuracy']:.3f}")

    # Example prediction demonstration
    print("\nExample Prediction:")
    new_song = pd.DataFrame({
        'danceability': [0.8],
        'energy': [0.6],
        'key': [5],
        'loudness': [-6.0],
        'mode': [1],
        'speechiness': [0.1],
        'acousticness': [0.2],
        'instrumentalness': [0.0],
        'liveness': [0.3],
        'valence': [0.7],
        'tempo': [120.0],
        'duration_ms': [200000]
    })

    # Making prediction for new song
    new_song_scaled = analyzer.scaler.transform(new_song)
    prediction = best_model.predict(new_song_scaled)

    # Printing prediction results
    print("\nInput features:")
    for feature, value in new_song.iloc[0].items():
        print(f"{feature}: {value}")
    print(f"\nPredicted cluster/popularity category: {prediction[0]}")

    # Showing probability predictions if available
    if hasattr(best_model, 'predict_proba'):
        probabilities = best_model.predict_proba(new_song_scaled)
        print("\nPrediction probabilities for each class:")
        for i, prob in enumerate(probabilities[0]):
            print(f"Class {i}: {prob:.3f}")
            
            # Every Team member contributed in all the Algorithm

if __name__ == "__main__":
    main()
```
